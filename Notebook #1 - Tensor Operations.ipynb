##Playing with NumPy:

````
a = np.array([1,2,3,4,5])
````

````
tensor_conversion = torch.from_numpy(a)
print(tensor_conversion)
print(tensor_conversion.type())
````

````
numpy_conversion = tensor_conversion.numpy()
print(numpy_conversion)
print(type(numpy_conversion))
````

##Playing with vectors:

````
t_one = torch.tensor([1,2,3])
t_two = torch.tensor([1,2,3])
print(t_one + t_two)
print(t_one * t_two)
print(t_one * 5)
dot_product = torch.dot(t_one,t_two)
print(dot_product)
````

````
x = torch.linspace(0,10, 100)
y = torch.exp(x)
````

````
plt.plot(x.numpy(),y.numpy())
````

````
sin_y = torch.sin(x)
````

````
plt.plot(x.numpy(), sin_y.numpy())
````

##Derivatives

````
x = torch.tensor(2.0, requires_grad = True)
y = 9*x**4 + 2*x**3 + 3*x**2 + 6*x + 1
y.backward()
x.grad
````

##Partial Derivatives

````
x = torch.tensor(1.0, requires_grad = True)
z = torch.tensor(2.0, requires_grad = True)
y = x**2 + z**3
y.backward()
print(x.grad)
print(z.grad)
````
